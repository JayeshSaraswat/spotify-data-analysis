# puting release date in main data frame using index

```python
df_tracks = pd.read_csv("D:\Spotify/tracks.csv")
df_tracks.set_index("release_date",inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

# to check the artist in 18th row in our dataset use index method

```python
df_tracks[["artists"]].iloc[18]
```

# convert mili seconds to seconds

```python
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True,axis = 1)
df_tracks.duration.head()
```

# 

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,16))
heatmap = sns.heatmap(corr_df,annot = True,fmt = ".1g",vmin =-1,center = 0,cmap = "inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklables(),rotation = 90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

# correaltion b/w loudness and energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y = "loudness", x = "energy",color = 'c').set(title="loudness vs energy correaltion")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y = "popularity", x = "acousticness",color = 'c').set(title="popularity vs acousticness correaltion")
```
